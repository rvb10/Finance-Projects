```python

def get_stock_returns(ticker: str, start_date: str, end_date: str, return_type: str = 'daily') -> pd.DataFrame:

    stock_data = yf.download(ticker, start=start_date, end=end_date)

    if stock_data.empty:
        raise ValueError(f"No data found for ticker {ticker} between {start_date} and {end_date}.")
    stock_data = stock_data[['Adj Close']]
    stock_data.rename(columns={'Adj Close': 'Price'}, inplace=True)

    if return_type == 'daily':
        stock_data['Returns'] = stock_data['Price'].pct_change()
    elif return_type == 'weekly':
        stock_data['Returns'] = stock_data['Price'].resample('W').ffill().pct_change()
    elif return_type == 'monthly':
        stock_data['Returns'] = stock_data['Price'].resample('M').ffill().pct_change()
    else:
        raise ValueError("Invalid return_type. Choose from 'daily', 'weekly', or 'monthly'.")
    stock_data.dropna(inplace=True)

    return stock_data
if __name__ == "__main__":
    tickers = temp_ticks
    start_date = start_date
    end_date = end_date
    file_name = "stock_returns.xlsx"
    return_type = "daily"

    fetch_returns_to_excel(tickers, start_date, end_date, file_name, return_type)
```

```python

def fetch_returns_to_excel(tickers, start_date, end_date, file_name="stock_returns.xlsx", return_type="daily"):
    all_returns = {}

    for ticker in tickers:
        try:
            print(f"Fetching data for {ticker}...")
            data = yf.download(ticker, start=start_date, end=end_date, progress=False)

            if not data.empty:
                data = data[['Close']]
                data.rename(columns={'Close': 'Price'}, inplace=True)
                if return_type == 'daily':
                    data['Returns'] = data['Price'].pct_change()
                elif return_type == 'weekly':
                    data['Returns'] = data['Price'].resample('W').ffill().pct_change()
                elif return_type == 'monthly':
                    data['Returns'] = data['Price'].resample('M').ffill().pct_change()
                else:
                    print(f"Invalid return_type for {ticker}. Skipping...")
                    continue
                all_returns[ticker] = data['Returns']
            else:
                print(f"No data found for {ticker}. Skipping...")
        except Exception as e:
            print(f"Failed to fetch data for {ticker}: {e}")
    if all_returns:
        returns_df = pd.DataFrame(all_returns)
        returns_df.index.name = "Date"
        returns_df.to_excel(file_name)
        print(f"Returns successfully exported to {file_name}.")
    else:
        print("No valid data to export.")


if __name__ == "__main__":
    tickers = temp_ticks
    start_date = start_date
    end_date = end_date
    file_name = "stock_returns.xlsx"
    return_type = "daily"

    fetch_returns_to_excel(tickers, start_date, end_date, file_name, return_type)
```
